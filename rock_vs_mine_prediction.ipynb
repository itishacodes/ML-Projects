Importing the dependencies.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and data processing.

```python
# loading the dataset to pandas dataframe.

sonar_data = pd.read_csv('/content/sonar data.csv', header=None)
```

```python
sonar_data.head() # this functions displays first five rows of our dataset.
```

```python
# number of rows and columns

sonar_data.shape
```

```python
# describes the data.
# describes --> statistical measures of the data.

sonar_data.describe()
```

```python
# finding the number of mines and rocks.

sonar_data[60].value_counts()

# 60 is the column number.
```

```python
# to get mean values.

sonar_data.groupby(60).mean()
```

```python
# separating data and labels.

X = sonar_data.drop(columns=60, axis=1) # dropping 60th column and storing the remaining in X.
Y = sonar_data[60] # storing 60th column in Y.
```

```python
print(X)
```

```python
print(Y)
```

Training and test data.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify = Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

MOdel training --> LOGISTIC REGRESSION MODEL


```python
model = LogisticRegression()
```

```python
# training the logistic regression model with training data.

model.fit(X_train, Y_train)
```

```python
# accuracy on training data.

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data: ', training_data_accuracy)
```

```python
# accuracy on test data.

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data: ', test_data_accuracy)
```
